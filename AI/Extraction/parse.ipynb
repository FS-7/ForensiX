```python
import pandas as pd
from datetime import datetime
```

```python

file = pd.read_csv("../Data/ex.csv", names=["Row", "Thread ID", "Address", "Person", "Date", "Date Sent", "Protocol", "Is Read", "Status", "Type", "Reply Address", "Subject", "Body", "Service Center", "Locked", "Sub ID", "Error Code", "Creator", "Seen"])
file.head()
```

```python
for i in file[1:]:
    file[i] = file[i].transform(lambda x: x.split("=")[1])
```

```python
file["Address"] = file["Address"].replace(regex="^\\+1", value="")
```

```python

file["Date"] = file["Date"].transform(lambda x: datetime.fromtimestamp(int(x)//1000))
```

```python
file["Date Sent"] = file["Date Sent"].transform(lambda x: datetime.fromtimestamp(int(x)//1000))
```

```python
#   TYPE: 1: INBOX, 2: SENT.
def parseType(x):
    print(x)
    if x == 1:
        return "Received"
    else:
        return "Sent"
file["Type"].transform(lambda x: parseType(x))
```
