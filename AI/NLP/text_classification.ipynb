USING GEMMA

```python
from transformers import pipeline
import torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float32 if torch.cuda.is_available() else torch.float16
gemma = "google/gemma-2-2b-it"
nlp = pipeline("text-generation", model=gemma, model_kwargs={"torch_dtype": torch_dtype}, device=device)

text = ""
messages = [
    {"role": "user", "content": 
        f"""
        {text}
        Query: Explain in one word classify the text into one of the following categories: Normal, Threat, identity_attack, sexual_explicit, toxicity, extremism
        """}
]

outputs = nlp(messages, max_new_tokens=128)
assistant_response = outputs[0]["generated_text"][-1]["content"].strip()
print(assistant_response)
```
