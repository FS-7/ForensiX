```python
import whisperx
import torch
import os

device = "cuda:0" if torch.cuda.is_available() else "cpu"
compute_type = "float16" if torch.cuda.is_available() else "int8"

batch_size = 12
language = "en"

audio_file = "test.m4a"

model = whisperx.load_model("large-v3", device=device, compute_type=compute_type, language=language)

results = []

loc = "./Data/"
audio_files = os.listdir(loc)

for audio in audio_files:
    audio = whisperx.load_audio(loc + audio_file)
    results.append(model.transcribe(audio, batch_size=batch_size))

for i, result in enumerate(results):
    for k in result["segments"]:
        print(f"{i}: {k['start']}\t- {k['end']}: {k['text']}")
```
