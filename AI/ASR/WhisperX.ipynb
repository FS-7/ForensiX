```python
import whisperx, torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.int8

asr = "openai/whisper-large-v3"
batch_size = 12
language = "en"
loc = "./Data/"
audio_file = "test.m4a"
```

```python
model = whisperx.load_model("large-v3", device, compute_type=str(torch_dtype).split(".")[1])
audio = whisperx.load_audio(loc+audio_file)
results = model.transcribe(audio, batch_size=batch_size, language='en')
```
