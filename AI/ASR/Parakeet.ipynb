```python
from nemo.collections.asr.models import SortformerEncLabelModel, ASRModel
```

```python
diar_model = SortformerEncLabelModel.from_pretrained(model_name="nvidia/diar_streaming_sortformer_4spk-v2")
asr_model = ASRModel.from_pretrained(model_name="nvidia/parakeet-tdt-0.6b-v2")

diar_model.eval()
```

```python
audio_file = "test_mono.m4a" #os.path.abspath("test.m4a")

result = diar_model.diarize(audio=audio_file, batch_size=16)
output = asr_model.transcribe([audio_file], timestamps=True)
print(result)
```

```python
#print(output[0])
segment_timestamps = output[0].timestamp['segment']
print(result)
print(segment_timestamps)
```

```python
print("Start\tEnd\tSpeaker\t\tText\n")
for text in segment_timestamps:
    print(f"{text['start']}\t{text['end']}: \t{"Speaker"}\t\t{text["segment"]}")
```
